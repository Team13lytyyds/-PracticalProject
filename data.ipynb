```python
import requests

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.125 Safari/537.36'}

res = requests.get('https://search.jd.com/Search?keyword=%E4%B9%A6%E7%B1%8D&enc=utf-8&wq=%E4%B9%A6%E7%B1%8D&pvid=78e0f64b03544d4db62ed27fab79eed7', headers = headers)
```

```python
print(res.text)
```

```python
import re
```

```python
name = re.findall('.*?class="p-name".*?<em.*?>(.*?)<.*?', res.text,  re.S)
```

```python
print(name)
```

```python
price = re.findall('.*?class="p-price".*?<i.*?>(.*?)</i>', res.text,  re.S)
```

```python
print(price)
```

```python
press = re.findall('.*?class="curr-shop hd-shopname".*?>(.*?)</a>', res.text,  re.S)
```

```python
print(press)
```

```python
import csv
```

```python
i=0
with open('jindongshuji.csv', 'w' , newline='', encoding='utf-8-sig') as csvfile:
    fieldnames=['标题','价格','出版社']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for item in zip(name,price,press):
        data = {}
        index=0
        for field in fieldnames:
            data[field] = item[index]
            index += 1
        writer.writerow(data)
        i+=1
        if i==10 :
            break
```
